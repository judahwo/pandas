### Iloc 

`iloc` is a property in the Pandas library used to access rows and columns of a DataFrame or Series by their integer-based index positions. Unlike label-based indexing (`loc`), `iloc` uses numerical indices (zero-based) to retrieve specific data.

```python
# Displaying row 90
df.iloc[90]
```

```python
# Displaying rows 90 to 95
df.iloc[90:95]
```

```python
# Displaying rows 90 to 95 and first two columns
df.iloc[90:95, 0:2]
```

```python
# Displaying rows 90 to 95 and two columns by name
df[['job_title_short', 'job_location']].iloc[90:95]
```

## Cleaning Dataframes

```python
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df.info()
```

Here we can see that the job_posted_date is an object. We need to convert it to a datetime object.<br>
**Documentation:** [pd.to_datetime()](https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html)


```python
# Retrieves the first value in the 'job_posted_date' column.
df.job_posted_date[0]
```

```python
# Checks the type of the first 'job_posted_date' value.
type(df.job_posted_date[0])
```

```python
# Converts the 'job_posted_date' column to datetime format.
pd.to_datetime(df.job_posted_date)
```

```python
df.info()
```

It's still an object

```python
# Converts 'job_posted_date' to datetime.
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df.info()
```

Now it is a datetime object. And we now have access to a new property.

```python
# Extracts the month from 'job_posted_date'.
df.job_posted_date.dt.month
```

```python
df.info()
```

But, we can see that our month value is not a part of our dataframe. We want to get it into our dataframe.

```python
# Adds a 'job_posted_month' column with the month from 'job_posted_date'.
df['job_posted_month'] = df.job_posted_date.dt.month
```

```python
df.info()
```

`job_posted_date.dt.month` is a now a part our dataframe.

```python
df
```

### Sorting Dataframes

```python
# Sorts the DataFrame by 'job_posted_date' in ascending order and updates it in place.
df.sort_values('job_posted_date', inplace=True)
```

```python
df
```

### Drop Method

**Documentation:** [pandas.DataFrame.drop](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html)


```python
# Drops the 'salary_hour_avg' column from the DataFrame.
df.drop(labels='salary_hour_avg', axis=1)
```

Here we can that that `salary_hour_avg` no longer shows up. But, if we run info, it is still in our dataframe.

```python
df.info()
```

```python
# Removes the 'salary_hour_avg' column from the DataFrame in place.
df.drop(labels='salary_hour_avg', axis=1, inplace=True)
```

```python
df.info()
```

And now we see that `salary_hour_avg` is no longer in our dataframe.

### Dropna Method

**Documentation:** [pandas.DataFrame.dropna](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html)

```python
# Drops rows where 'salary_year_avg' has missing values.
df.dropna(subset=['salary_year_avg'])
```

Here we can see that we have dropped all rows with NaN values in `salary_year_avg`. But, again, if we run info, we can see that the rows are still there and weren't filtered out of the dataframe. We can see that there are over 700K rows in our dataframe, but only 22K entries with non-null values in `salary_year_avg`.

```python
df.info()
```

```python
# Removes rows with missing 'salary_year_avg' values from our dataframe.
df.dropna(subset=['salary_year_avg'], inplace=True)

```

```python
df.info()
```

Now we can see that we have dropped all rows with NaN values in `salary_year_avg` and that the rows are no longer in our dataframe.

## Analyzing Dataframes

```python
import pandas as pd
from datasets import load_dataset

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.describe()
```

```python
# Gets the median of 'salary_year_avg'
float(df['salary_year_avg'].median())
```

```python
# Gets the minimum of 'salary_year_avg'
float(df['salary_year_avg'].min())
```

```python
# Gets the index of the minimum 'salary_year_avg'
float(df['salary_year_avg'].idxmin())
```

```python
# Finds the index of the row with the minimum 'salary_year_avg'.
min_salary = df['salary_year_avg'].idxmin()
```

```python
# Retrieves the row corresponding to the minimum 'salary_year_avg'.
df.iloc[min_salary]
```

```python
# Lists unique values in the 'job_title_short' column.
df['job_title_short'].unique()
```

```python
# Counts the occurrences of each unique value in 'job_title_short'.
df['job_title_short'].value_counts()
```

### pandas.DataFrame.groupby

**Documentation:** [pandas.DataFrame.groupby](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html)

```python
# Finds the minimum 'salary_year_avg' for each 'job_title_short' group.
df.groupby('job_title_short')['salary_year_avg'].min()
```

```python
# Calculates the median 'salary_year_avg' for each 'job_title_short' group.
df.groupby('job_title_short')['salary_year_avg'].median()

```

```python
# Calculates the median of 'salary_year_avg' and 'salary_hour_avg' for each 'job_title_short' group.
df.groupby('job_title_short')[['salary_year_avg', 'salary_hour_avg']].median()

```

```python
# Computes the min, max, and median of 'salary_year_avg' for each 'job_title_short' group.
df.groupby('job_title_short')['salary_year_avg'].agg(['min', 'max', 'median'])
```
